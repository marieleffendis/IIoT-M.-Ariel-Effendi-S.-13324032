```python
from ultralytics import YOLO
import os
```

```python
DATASET_PATH = r"D:\IIOT\dataset"
DATA_YAML = os.path.join(DATASET_PATH, "data.yaml")
```

```python
MODEL_BASE = "yolo26s.pt"   # model ringan (recommended)
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 8
DEVICE = "cpu"

print("Config loaded!")
```

```python
# ==============================
# CEK FILE
# ==============================
print("data.yaml exists:", os.path.exists(DATA_YAML))

train_img_path = os.path.join(DATASET_PATH, "train/images")
val_img_path = os.path.join(DATASET_PATH, "val/images")

print("Train images:", len(os.listdir(train_img_path)))
print("Val images:", len(os.listdir(val_img_path)))
```

```python
# ==============================
# LOAD MODEL
# ==============================
model = YOLO(MODEL_BASE)

print("Model loaded!")
```

```python
# ==============================
# TRAINING
# ==============================
results = model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    device=DEVICE,
    name="training_box_detector",
    verbose=True
)
```

```python
# ==============================
# PATH MODEL HASIL
# ==============================
MODEL_PATH = r"D:\IIOT\runs\detect\training_box_detector-2\weights\best.pt"

print("Model saved at:", MODEL_PATH)
```

```python
# ==============================
# TEST INFERENCE
# ==============================
model = YOLO(MODEL_PATH)

test_image = os.path.join(DATASET_PATH, "val/images", os.listdir(val_img_path)[0])

results = model(test_image, show=True)
```
